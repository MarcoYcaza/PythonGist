```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("nuforc_reports.csv")
```

```python
df
```

```python
df.info()
```

```python
df.duration.fillna("",inplace=True)
```

```python
df
```

```python
df["naming_length"] = df.duration.apply(lambda valor :  len(valor))
```

```python
df=df[["summary","city","state","shape","duration","naming_length"]]
```

```python
sns.countplot(df.naming_length)
```

```python
df.loc[df.naming_length == 10]          
```

```python
temp = df.duration.str.extract('(\d+) (\w+)',expand=True)
temp.columns = ["time","scale"]

df=df.join(temp,on=df.index)

df.scale = df.scale.str.lower()
df.scale.fillna("",inplace=True)
```

```python
df.head()
```

```python
df.scale.unique()
```

```python
#df.loc[df.scale=="seoonds"]
```

```python
df["seconds"] = np.where(df.scale.str.contains("min"),pd.to_numeric(df.time)*60,pd.to_numeric(df.time))

df["seconds"] = np.where(df.scale.str.contains("h"),pd.to_numeric(df.seconds)*3600,pd.to_numeric(df.seconds))
```

```python
df
```

```python
df.iloc[88065]
```

```python
df.fillna(0,inplace=True)
df.sort_values(by="seconds",ascending=True,inplace=True)
df.reset_index(drop=True)
```

```python
df_final = df.loc[(df.seconds!=0)]
```
